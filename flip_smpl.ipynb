```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'demo_walk_9_train'
data = pd.read_csv(f'train/{file}.csv', index_col=0)
```

```python
#get all columns that contain the word 'left'
left_columns = [col for col in data.columns if 'left' in col]
print(left_columns)

right_columns = [col for col in data.columns if 'right' in col]
print(right_columns)

data_flip = data.copy()

for left_col, right_col in zip(left_columns, right_columns):
    data_flip[right_col] = data[left_col]

for right_col, left_col in zip(right_columns, left_columns):
    data_flip[left_col] = data[right_col]
```

```python
data_flip.to_csv(f'train/{file}_flip.csv')
```

```python
#plot nth column of data and data_flip
n = 10
plt.plot(data.iloc[:,n])
plt.plot(data_flip.iloc[:,n])
plt.show()
```
